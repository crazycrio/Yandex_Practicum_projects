#### Убираем редкие и выбивающиеся значения

Пропишем функцию, которая будет находить межквартильный размах для каждого столбца и строить графики, для отфильтрованных данных. Которой будем пользоваться и дальше.

```python
def iqr_filter(df, column,plot=0,x=10,y=10, plot_kind = 'hist'):
    q1 = df[column].quantile(0.25)                 
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    filtered = (df[column] >= (q1 - 1.5*iqr)) & (df[column] <= (q3 + 1.5*iqr))
    if plot == 1:
        return df.loc[filtered].plot(y = [column], figsize=(x,y), kind = plot_kind, grid = True)
    else:
        return df.loc[filtered]
```

```python
cols_to_analise = ['total_area_m2', 'last_price', 'number_of_rooms', 'ceiling_height_m']

for col in cols_to_analise:

    iqr_filter(df, col, 1)
```

#### Оценка влияния факторов на цену квартиры в центре города и за его пределами

Как ранее было выявлено, граница центральной зоны Санкт-Петербурга проходит на отметке 8км. Сделаем два среза из таблицы по этому значению, квартиры в центре и вне центра, посчитаем для каждой корреляции между параметрами, затем соединим это в одной таблице, чтобы наглядно представить сравнение корреляции каждого параметра с финальной ценой для квартир в центре и вне центра.

```python
corr_in_central_area = df_piter.query('center_distance_km <= 8').corr().loc['last_price'].sort_values(ascending = False).reset_index()
corr_in_central_area.style.background_gradient('coolwarm')
```

```python
corr_out_central_area = df_piter.query('center_distance_km > 8').corr().loc['last_price'].sort_values(ascending = False).reset_index()
corr_out_central_area.style.background_gradient('coolwarm')
```

```python
total_correlation = corr_in_central_area.merge(corr_out_central_area, on='index')
total_correlation = total_correlation.rename(columns={'last_price_x':'cor_in_central_area', 'last_price_y':'cor_out_central_area'})
total_correlation.style.background_gradient('coolwarm')
```

**Вывод**

Главные факторы влияющие на цену квартиры в центре и за пределами это количество комнат, цена за квадратный метр и площадь. Но влияют они по разному. Для центра факторы расположились в таком порядке: цена за квадрат, общая площадь, число комнат. Вне центра главный фактор - общая площадь, затем число комнат и почти наравне с ним идет цена за квадрат.

В целом вывод не противоречат житейской логике - в центре продаются квартиры меньше и с меньшим числом комнат, но за счет центральности, растет цена за квадратный метр. На перефирии же наоборот, обосновываются люди чаще семейные, поэтому больший фактор имеет размер квартиры, количество комнат и цена за квадрат влияют почти одинаково.
